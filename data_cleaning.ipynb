```python
postprocessed_data = pd.read_csv('data/wikilarge/intermediate/wiki_combined_labelled.csv')
print(f"Postprocessed data shape: {postprocessed_data.shape}")
print(postprocessed_data.columns)
print(postprocessed_data.head(10))
```

```python
# filter out rows with label 0
filtered_data = postprocessed_data[postprocessed_data['label'] != 0].reset_index(drop=True)
print(f"Filtered data shape: {filtered_data.shape}")
filtered_data.head(10)
print(filtered_data.columns)
# copy filtered data and flip src and tgt columns (and respective info columns) to double the dataset
```

```python
src_columns = [column for column in filtered_data.columns if 'src' in column]
tgt_columns = [column for column in filtered_data.columns if 'tgt' in column]
print(f"Source columns: {src_columns}")
print(f"Target columns: {tgt_columns}")
print(f"number of source columns: {len(src_columns)}")
print(f"number of target columns: {len(tgt_columns)}")

# copy filtered data and flip src and tgt columns
flipped_data = filtered_data.copy()
for src_col, tgt_col in zip(src_columns, tgt_columns):
    flipped_data[src_col], flipped_data[tgt_col] = flipped_data[tgt_col], flipped_data[src_col]
print(f"Flipped data shape: {flipped_data.shape}")
flipped_data.head(10)
```

```python
# concat filtered_data and flipped_data
doubled_data = pd.concat([filtered_data, flipped_data], ignore_index=True)
print(f"Final data shape: {final_data.shape}")
```

```python
doubled_data.to_csv('data/wikilarge/intermediate/wiki_cleaned_doubled.csv', index=False)
```
